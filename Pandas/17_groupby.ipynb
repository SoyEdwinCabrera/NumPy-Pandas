```python
import pandas as pd
```

```python
df_books = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/j.csv', sep=',',header=0)
df_books.head(2)
```

```python
df_books.groupby('Author').count()
```

```python
numerical_column = df_books.select_dtypes(include=['number']).columns
df_books.groupby('Author')[numerical_column].mean()
```

```python
df_books.groupby('Author')[df_books.select_dtypes(include='number').columns].mean()
```

```python
df_books.groupby('Author').min()
```

```python
df_books.groupby('Author').max()
```

```python
df_books.groupby('Author').sum()
```

```python
df_books.groupby('Author').agg(['min','max'])
```

```python
df_books.groupby('Author').agg({'Reviews':['min','max'],'User Rating':'sum'})
```
