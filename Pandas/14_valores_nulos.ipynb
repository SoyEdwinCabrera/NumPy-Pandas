```python
import pandas as pd
import numpy as np
```

```python
dict = {'Col1':[1,2,3,np.nan],
        'Col2':[4,np.nan,6,7],
        'Col3':['a','b','c',None]}
```

```python
df = pd.DataFrame(dict)
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull()*1
```

```python
df.fillna('Missing')
```

```python
df.fillna(df.mean())
```

```python
df.fillna({"Col1":df["Col1"].mean()}, inplace=True)
df
```

```python
df = df.fillna(df.select_dtypes(include=np.number).mean())
df
```
